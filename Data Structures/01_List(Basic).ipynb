### Definition  
A list is an ordered, mutable collection in Python.  
It can store multiple items and different data types.

Example:[1,2,3]


#### Features of Lists
- Ordered  
- Mutable (can be changed)  
- Allows duplicate values  
- Supports indexing & slicing  


```python
# Program 1: Creating and accessing list elements
my_list = [10, 20, 30, 40, 50]
print("List ::",my_list)
print('Type ::',type(my_list))
```

```python
# 2. accessing list elements 
print("First element:", my_list[0])
print("Last element:", my_list[-1])
```

```python
# 3.update the list
my_list[2]=45
my_list[3]=54
print(my_list)
```

```python
# 4.length of list 
print("Length =", len(my_list))
```

```python
# 5. Slicing
nums = [1, 2, 3, 4, 5, 6, 7]

print("Original:", nums)
print("First 3 =", nums[:3])
print("Middle 3 =", nums[2:5])
print("Every 2nd element =", nums[::2])
print("Reverse list =", nums[::-1])
```

```python
# 6. Basic operations
a = [1, 2, 3]
b = [4, 5]
print("Concatenation:", a + b)
print("Repetition:", a * 2)
print("Is 2 in list a?", 2 in a)
```

```python
# 7. Unpacking of list 
list=['ramesh','ram']
a,b=list
print(a)
print(b)
```
