## Topic : Data Structures – Dictionary
### Definition  
A dictionary is an unordered collection of data stored in **key–value pairs**.  
Each key must be unique, and values can be of any data type.



```python
# Creating and accessing dictionary

student = {
    "name": "Bhavesh",
    "age": 20,
    "branch": "AI & DS"
}

print("Student Dictionary:", student)
print("Name:", student["name"])
print("Age:", student.get("age"))
```

```python
# Update dictionary
student = {
    "name": "Bhavesh",
    "age": 21
}
# Add new key
student["city"] = "Nagpur"
# Update existing key
student["age"] = 22
print("Updated Dictionary:", student)
```

```python
var={'Name':'Bhavesh','Age':21,'cgpa':9.2}
print(var)
```

```python
#value in list
var={'Name':'Bhavesh','Age':21,'cgpa':9.2}
print(var['Age'])
print(var['Name'])
print(var['cgpa']) 
```

```python
#pop method
var={'Name':'Mangesh','Age':21,'cgpa':8.4}
var.pop('Age')
print(var) 
```

```python
#get
var={'Name':'Mangesh','Age':21,'cgpa':8.4}
print(var.get('Name')) 
print(var.get('pass',"Not avalible")) 
```

```python
#clear
var={'Name':'Mangesh','Age':21,'cgpa':8.4}
var.clear()
print(var)
```

```python
var={'Name':'Mangesh','Age':21,'cgpa':8.4}
for key,values in var.items():
    print(key,values,sep="-")
```

```python
#Update
var={'Name':'Mangesh','Age':21,'cgpa':8.4}
var['Age']=22
print(var)
```

```python
var={'Name':'Mangesh','Age':21,'cgpa':8.4}
print("Key values::",var.keys())
print("values::",var.items()) 
```
