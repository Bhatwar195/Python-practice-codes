## Topic: Set in Python 
###  Definition  
A set is an unordered collection of unique elements in Python.  
Sets do not allow duplicate values and are mainly used for storing distinct data.

Syntax:{1,3,5}

```python
# Creating a set and duplicate removal
data = {1, 2, 2, 3, 4, 4, 5}
print("Set:", data)
```

```python
#Empty set 
set1={}
print(type(set1)) 
#It not show class set it show class dict
```

```python
# Check the elemnt in set 
fruits = {"apple", "banana", "mango"}

print("apple in set?", "apple" in fruits)
print("grapes in set?", "grapes" in fruits)
```

```python
# add(),remove(),discard(),pop(),clear()
fruits = {"apple", "banana", "mango"}

fruits.add("orange")
fruits.remove("banana")
fruits.discard("grapes")   # no error if element not present
fruits.pop()
print("Updated set:", fruits)
fruits.clear()
print("After clear:",fruits)
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("Union:", A | B)
print("Intersection:", A & B)
print("Difference (A - B):", A - B)
print("Symmetric Difference:", A ^ B)
```
