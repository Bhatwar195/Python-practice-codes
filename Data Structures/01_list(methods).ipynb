## Topic :  List Methods & Advanced Programs
### Important List Methods
- append()  
- insert()  
- remove()  
- pop()  
- sort()  
- reverse()  

```python
# Append
fruits = ["apple", "banana", "mango"]
fruits.append("orange")
print(fruits)
```

```python
# Insert
fruits = ["apple", "banana", "mango"]
fruits.insert(1, "grapes")
print(fruits)
```

```python
# Remove
fruits = ["apple", "banana", "mango"]
fruits.remove("banana")
print(fruits)
```

```python
# Pop
fruits = ["apple", "banana", "mango"]
last_item = fruits.pop()   # removes last
print("Final List:", fruits)
print("Popped item:", last_item)
```

```python
# Program 5: Sorting and reversing
marks = [50, 90, 70, 20, 80]
marks.sort()
print("Sorted:", marks)

marks.reverse()
print("Reversed:", marks)
```

```python
# List comprehension to create squares
nums = [1, 2, 3, 4, 5]
squares = [n*n for n in nums]

print("Numbers:", nums)
print("Squares:", squares)
```
