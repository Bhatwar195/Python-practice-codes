## Topic :Tuples 
### Definition  
A tuple is an ordered and immutable collection of elements in Python.  
Once created, a tuple cannot be modified.

Tuples are written using parentheses `( )`.

Example:(1,2,3,4)


```python
# Program 1: Creating and accessing tuple elements
my_tuple = (10, 20, 30, 40, 50)

print("Tuple:", my_tuple)
print("First element:", my_tuple[0])
print("Last element:", my_tuple[-1])
print("Slice 1:4 =", my_tuple[1:4])
print("Length =", len(my_tuple))
```

```python
# Tuple operations
a = (1, 2, 3)
b = (4, 5)

print("Concatenation:", a + b)
print("Repetition:", a * 2)
print("Is 2 in a?", 2 in a)
```

```python
#  Tuple slicing
t = (1, 2, 3, 4, 5, 6)

print("Original:", t)
print("First 3:", t[:3])
print("Middle:", t[2:5])
print("Reverse:", t[::-1])
```

```python
# find the length of tuple
var = (1,2,3,4,True,False,"Python","Like")
print(len(var))
```

```python
# copy one tuple variable to another variable
var = (1,2,3,4,True,False,"Python","Like")
var2 = var
print(var2)
print(len(var2)) 
```

### Features of Tuples
- Ordered  
- Immutable (cannot be changed)  
- Allows duplicate values  
- Faster than lists  
- Used for fixed data
