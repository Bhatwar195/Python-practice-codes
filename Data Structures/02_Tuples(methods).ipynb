## Topic : Tuples methods 

```python
# count() and index()
t = (10, 20, 30, 20, 40, 20)

print("Tuple:", t)
print("Count of 20:", t.count(20))
print("Index of 30:", t.index(30))
```

```python
# Tuple packing and unpacking
# Packing
student = ("Bhavesh", 19, "AI & DS")
# Unpacking
name, age, branch = student
print("Name:", name)
print("Age:", age)
print("Branch:", branch)
```

```python
#search element in tuple 
fruit_tuple = ("Apple", "Orange", "Mango")
str=input("Enter any string to search:")
if str in fruit_tuple:
 print(str," is found")
else:
 print("Not found")
```

```python
#  Nested tuple
records = (
    ("A", 90),
    ("B", 85),
    ("C", 88)
)

for r in records:
    print("Name:", r[0], "Marks:", r[1])
```
