## Topic: Python Basics – Operators
### Definition  
Operators are symbols used to perform operations on variables and values in Python.

### Types of Operators:
1. **Arithmetic Operators** → +, -, *, /, %, //, **  
2. **Relational / Comparison Operators** → ==, !=, >, <, >=, <=  
3. **Logical Operators** → and, or, not  
4. **Assignment Operators** → =, +=, -=, *=, /=  
5. **Membership Operators** → in, not in  
6. **Identity Operators** → is, is not


```python
# Program 1: Demonstration of arithmetic operators
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
```

```python
# Program 2: Comparison (Relational) Operators
x = 15
y = 20

print("x == y:", x == y)
print("x != y:", x != y)
print("x > y:", x > y)
print("x < y:", x < y)
print("x >= y:", x >= y)
print("x <= y:", x <= y)
```

```python
# Program 3: Logical Operators
age = 21
is_student = True

print("age > 18 and is_student:", age > 18 and is_student)
print("age > 18 or is_student:", age > 18 or is_student)
print("not is_student:", not is_student)
```

```python
# Program 4: Assignment Operators

num = 10
print("Initial:", num)

num += 5
print("After += 5:", num)

num -= 3
print("After -= 3:", num)

num *= 2
print("After *= 2:", num)

num /= 4
print("After /= 4:", num)
```

```python
# Program 5: Membership and Identity Operators
items = [10, 20, 30, 40]
a = 20
print("20 in items:", 20 in items)
print("50 not in items:", 50 not in items)

x = [1, 2, 3]
y = [1, 2, 3]

print("x == y :", x == y)      # values same
print("x is y :", x is y)      # identity check (False)
print("x is not y:", x is not y)
```
