### Definition  
Lambda functions are small, anonymous (nameless) functions.  
They are used when a function is needed for a short duration.

### Syntax:
lambda arguments : expression

```python
# 1.simple lambda function

square = lambda x: x * x
print("Square of 6 ::", square(6))
```

```python
# 2.Lambda function with two argument

add = lambda a, b: a + b
print("Addition =", add(10, 5))
```

```python
# 3. Check even or Odd
even=lambda num:num%2==0
even(31)
```

```python
# 4.lambda function with multiple parameter 
Addtion=lambda x,y,z:x+y+z
```

```python
Addtion(8,9,10)
```

### Conclusion
By using lambda functions, we can perform operations concisely without defining full functions.
Lambda functions are especially powerful when used with higher-order functions like map and filter.
