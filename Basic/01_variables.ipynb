## Topic: Variables In Python
Variables are used to store data in Python.
They do not need explicit type declaration, and their type depends on the value assigned.
Python variables are dynamically typed and easy to update.

```python
# simple variable assignment
name = "Bhavesh"
age = 21
score = 85.5
```

```python
# printing variables
print("Name:", name)
print("Age:", age)
print("Score:", score)
```

```python
# dynamic typing (variable type can change anytime)
age = "Twenty One"   # changing integer to string
print("Updated Age:", age)
```

```python
# multiple variable assignment
x, y, z = 10, 20, 30
print("x:", x, "y:", y, "z:", z)
```

```python
# swapping variables (Python trick)
a, b = 5, 10
a, b = b, a
print("After swap -> a:", a, "b:", b)
```
