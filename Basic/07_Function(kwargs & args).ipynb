## Topic : Part 5 Advanced Function Concepts
This part includes:

- Default arguments  
- Keyword arguments  
- *args (variable length positional arguments)  
- **kwargs (variable length keyword arguments)  

```python
#1.Default argument 
def greet(name="Guest"):
    print("Hello,", name)

greet()
```

```python
def greet(name="Guest"):
    print("Hello,", name)
    
greet("Bhavesh")
```

```python
# keyword Argument 
def student(name, age):
    print("Name:", name)
    print("Age:", age)

student(age=21, name="Bhavesh")
```

```python
# *args
def total(*numbers):
    print("Sum =", sum(numbers))

total(1, 2, 3, 4, 5)
```

```python
# **kwargs
def details(**info):
    for key, value in info.items():
        print(key, ":", value)

details(name="Bhavesh", branch="AI & DS", year=1)
```

```python
# combining args and kwargs 
def print_all(*args,**kwargs):
    print('Positional arguments :')
    for val in args:
        print(val)
    print("Keyword arguments:")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_all(1,2,3,"end",name="Bhavesh",Age="20",city="nagpur")
```
