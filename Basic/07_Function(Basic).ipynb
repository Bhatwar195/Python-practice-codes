### Definition  
A function is a reusable block of code that performs a specific task.  
Functions help in writing cleaner, modular and efficient programs.

```python
# 1 Simple function
def greet():
    print("Hello! Welcome to Python Functions.")

greet()
```

```python
# 2 Function With Argument
def add(a, b):
    return a + b

print("Sum =", add(10, 20))
```

```python
# 3 Function With returning values 
def get_name():
    return "Bhavesh"

print("Name:", get_name())
```

```python
# 4 Types of Function based on Parameters

# 1 No argument, no return
def show():
    print("Hello")

# 2 Argument, no return
def display(name):
    print("Welcome", name)

# 3 No argument, return value
def get_pi():
    return 3.14

# 4 Argument + return value
def multiply(a, b):
    return a * b

show()
display("Bhavesh")
print(get_pi())
print("Multiplication =", multiply(4, 5))
```

### Conclusion  
Simple functions help reduce repetition and make code easy to understand.  
These are the foundation for advanced function concepts.
