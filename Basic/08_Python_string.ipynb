## Topic : Python string 
### Definition  
A string is a sequence of characters enclosed in quotes (`" "` or ' ').  

```python
text = "Python Programming"

print("Original String:", text)
print("Length:", len(text))
print("First character:", text[0])
print("Last character:", text[-1])
print("Substring (0 to 6):", text[0:6])
print("Repeated:", text * 2)
print("Contains 'Pro'? ->", "Pro" in text)
```

```python
text = "Hello Python"

print("Uppercase:", text.upper())
print("Lowercase:", text.lower())
print("Starts with 'Hello'? ->", text.startswith("Hello"))
print("Replace Python -> World:", text.replace("Python", "World"))
print("Split:", text.split())
print("Find 'Py':", text.find("Py"))
```

```python
word = "Programming"

print("Original:", word)
print("Slice 0:5 :", word[0:5])
print("Every 2nd character:", word[::2])
print("Reverse:", word[::-1])
```

```python
### String Formatting (f-strings)
name = "Bhavesh"
age = 19
print(f"My name is {name} and I am {age} years old.")
```

```python
### String formating 
name = "Bhavesh"
course = "Machine Learning "
print(f"My name is {name} and I learn {course}.")
```

```python
# Palindrome Series 
word = "madam"

if word == word[::-1]:
    print(word, "is a palindrome")
else:
    print(word, "is not a palindrome")
```

```python
## Remove space 
sentence = "Machine learning"
new_sentence = sentence.replace(" ", "")

print("Original:", sentence)
print("Without spaces:", new_sentence)
```
