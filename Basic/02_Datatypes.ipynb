## Topic : Data Types in Python
#### Definition  
Data Types represent the type of value stored in a variable.  
Python supports different built-in data types such as:
- Integer  
- Float  
- String  
- Boolean  
- List  
- Tuple  
- Set  
- Dictionary  

Python is dynamically typed, meaning the data type is assigned automatically based on the value.


```python
# Integer
age = 21
print("Integer:", age)
print('Data type of age :',type(age))
```

```python
# Float
percentage = 85.5
print("Float:", percentage, type(percentage))
```

```python
# String
name = "Alice"
print("String:", name, type(name))
```

```python
# Boolean
is_student = True
print("Boolean:", is_student, type(is_student))
```

### Advanced Datatypes

```python
# List
marks = [85, 90, 75, 95]
print("List:", marks, type(marks))
```

```python
# Tuple
coordinates = (10, 20)
print("Tuple:", coordinates, type(coordinates))
```

```python
# Set
unique_nums = {1, 2, 3, 3, 2}
print("Set:", unique_nums, type(unique_nums))
```

```python
# Dictionary
student = {
    "name": "Bhavesh",
    "age": 21,
    "branch": "AI & DS"
}
print("Dictionary:", student, type(student))
```

## Conclusion  
In this program, we explored Python's commonly used data types.  
Each data type serves a different purpose:

- **int** → whole numbers  
- **float** → decimal values  
- **str** → text data  
- **bool** → True/False  
- **list** → ordered, changeable collection  
- **tuple** → ordered, unchangeable collection  
- **set** → unique unordered values  
- **dict** → key-value pairs  

Understanding data types is essential because they determine how Python stores and processes data.
