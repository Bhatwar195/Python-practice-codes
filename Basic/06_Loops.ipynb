## Topic : Python Basics – Loops (for & while)
### Definition  
Loops are used to repeat a block of code multiple times.

### Types of Loops:
1. **for loop** → used to iterate over a sequence (list, string, range, etc.)  
2. **while loop** → runs until a condition becomes False  

Loops help in automating repetitive tasks efficiently.


```python
# Program 1: Print numbers from 1 to 10 using a for loop
for i in range(1, 11):
    print(i)
```

```python
# Program 2: Countdown from 5 to 1 using while loop
n = 5
while n > 0:
    print(n)
    n -= 1
```

```python
# Program 3: Calculate sum of first 10 natural numbers

total = 0
for i in range(1, 11):
    total += i

print("Sum of 1 to 10 =", total)
```

```python
# Program 4: Iterate through a list and print items

fruits = ["apple", "banana", "mango", "orange"]

for fruit in fruits:
    print("Fruit:", fruit)
```

```python
# Program 5: Demonstrate break and continue
numbers = [1, 2, 3, 4, 5, 6]

print("Using continue (skip 3):")
for n in numbers:
    if n == 3:
        continue
    print(n)

print("\nUsing break (stop at 4):")
for n in numbers:
    if n == 4:
        break
    print(n)
```

```python
# Print string multiple time 
i = 0
while i < 5: 
    print("Bhavesh") 
    i = i + 1
```

## Conclusion  
Loops help in executing repeated tasks efficiently.

### Key Points:
- `for` loop → best for iterating over sequences  
- `while` loop → best when number of iterations is unknown  
- `break` → stops the loop immediately  
- `continue` → skips only the current iteration
