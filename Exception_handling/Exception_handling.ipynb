## Topic : Exception Handling 
### Definition  
Exception handling is used to handle runtime errors so that the normal flow of the program does not stop.

Python uses:
- `try`
- `except`
- `else`
- `finally`

to handle exceptions safely.


```python
# try,except
try:
    print("value of a :",a)
except:
    print("value of a is not define") 
```

```python
# Handling multiple exceptions
try:
    x = int("abc")
    y = 10 / 0
except ValueError:
    print("ValueError occurred")
except ZeroDivisionError:
    print("ZeroDivisionError occurred")
```

```python
#try,except,else
x=20
y=0
try:
    print("devide::",x/y)
except:
    print("can not devide by zero")
else:
    print("end progran")
```

```python
#try,except,else,finaly
x=20
y=0
try:
    print("devide::",x/y)
except Exception as e:
    print(e)
else:
    print("It is a else block")
finally:
    print("It is finally block") 
```

```python
#  finally block
try:
    f = open("test.txt", "r")
    print(f.read())
except FileNotFoundError:
    print("File not found")
finally:
    print("Closing file operation")
```

```python
#Raising exception manually
age = -5

if age < 0:
    raise ValueError("Age cannot be negative")
```
